```python
def show_dataframe_info(path):
    df = pd.read_csv(path)
    print(df.count())
    display(df.head())

for file in os.listdir('./csv_files'):
    p = os.path.join(os.getcwd(), 'csv_files', file)
    print(file.split('.')[0])
    show_dataframe_info(p)
```
